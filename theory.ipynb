## Difference between Time Series Model & Structural Model

`Time Series Model`
- Model variable based on the variable's past value

`Structural Model`
- Model variable based on some functions based on arguments 
- Linear Regression is example of structural model

`Why use Time Series Analysis in finance forecasting?`
- A lot of data available on daily basis
- Not all factors are observable or obtainable to use regression easily
- It is alternative forecasting technique

## Stationary Process - Strong Stationary & Weak Stationary

Time Series are **random processes**, they have **non-deterministic** nature.

`Joint distribution` is 2 or more distributions combined. 
- For example *P1 = P(y<sub>t</sub> = 2, y<sub>t+1</sub> =3)*

`Strong distribution` is when **joint distribution** of time series remains unchanged over time.
- For example, taking some joint distributions *P1, P2, P3* it should be true that *P1=P2=P3*. It should be true for **all** data points

`Weak Stationary` is when **mean** of time series remains unchanged over time, **variance** and **autocovariance** stays the same as well.

**Autocovariance** depends only on time so for all given delta t **Autocovariance** of y<sub>t</sub> and y<sub>t-delta</sub> are **equal**. It is special type of covariance because two series come from same series, hence name **autocovariance**.

All **strong stationary** are also **weak stationary** but not the other way around.

**Weak Stationary** is most likely to be found in real life scenarios.

## Auto Correlaction Function in Time Seriess Analysis

`ACF` is basically **Auto Covariance** divaded by **Variance**.
It is also known as `correlogram`. **ACF** is also normalized.
If S<sub>t</sub> is correlated to S<sub>t-1</sub> this means it is **first order** dependency. Ideally we want S<sub>t</sub> to be strongly correlated to **first lag** and less and less with futher **lags**. **ACF** generally behaves likes correlation function.

S<sub>t-1</sub> is `lag` of **order 1** of S<sub>t</sub>.
- For example S<sub>t</sub> *= {1,2,3,4,5,...}* so basically **all data points**. <br>Then S<sub>t-1</sub> *= {2,3,4,5,6,...}* so all data points **expect first one**. It can be also written as S<sub>t-1</sub> = L(S<sub>t</sub>).

`Auto Covariance` is calculated as: <br>
*E[(S<sub>t</sub> - M<sub>t</sub>)(S<sub>t-1</sub> - M<sub>t-1</sub>)]*

## Partial Auto Correlation Function 

`PACF` is conditional correlation. It takes out effect of all intermediate lags. It is useful in detecting order of **AR Process**.


## Auto Regressive TIme Series Model

`AR` model depends of **past values** and **error term**.
**Stationary condition** matter a lot in **AR**. <br>
**Non stationary** series will have **non declining** effect which is not a good thing. Without testing stationary you can't do estimation.
